## Activation Functions

### ReLU Function

```python
x = jnp.arange(-8.0, 8.0, 0.1)
y = jax.nn.relu(x)
d2l.plot(x, y, 'x', 'relu(x)', figsize=(5, 2.5))
```

```python
y = jax.vmap(jax.grad(jax.nn.relu))(x)
d2l.plot(x, y, 'x', 'grad of relu', figsize=(5, 2.5))
```

### Sigmoid Function

```python
y = jax.nn.sigmoid(x)
d2l.plot(x, y, 'x', 'sigmoid(x)', figsize=(5, 2.5))
```

```python
y = jax.vmap(jax.grad(jax.nn.sigmoid))(x)
d2l.plot(x, y, 'x', 'grad of sigmoid', figsize=(5, 2.5))
```

## Tanh Function

```python
y = jax.nn.tanh(x)
d2l.plot(x, y, 'x', 'tanh(x)', figsize=(5, 2.5))
```

```python
y = jax.vmap(jax.grad(jax.nn.tanh))(x)
d2l.plot(x, y, 'x', 'grad of tanh', figsize=(5, 2.5))
```
